```python
%store -r Z_GF
%store -r Z_QL
%store -r Z_PG
%store -r Z_CF
%store -r Z_KC

Z_GF1 = np.array(Z_GF)//int(9)
Z_QL1 = np.array(Z_QL)//int(9)
Z_PG1 = np.array(Z_PG)//int(9)
Z_CF1 = np.array(Z_CF)//int(9)
Z_KC1 = np.array(Z_KC)//int(9)
```

```python
M_odot = 2.08

def Rkm(rho_c,varkappa):
    return np.round(((M_odot*Msol/4/np.pi/rho_c/(1/3 - (1-varkappa)/5))**(1/3))/1000,2)

Rkms = []

for i in range(len(Lista_rho_c)):
    
    for j in range(len(Lista_varkappa)):
        
        if Rkm(Lista_rho_c[i],Lista_varkappa[j]) <= 13.1 and Rkm(Lista_rho_c[i],Lista_varkappa[j]) >= 11.6:
            
            Rkms.append(int(1))
        
        else: Rkms.append(int(0))
       
```

```python
NumNICER_GF = [i.sum() for i in (Z_GF1*Rkms)]
NumNICER_QL = [i.sum() for i in (Z_QL1*Rkms)]
NumNICER_PG = [i.sum() for i in (Z_PG1*Rkms)]
NumNICER_CF = [i.sum() for i in (Z_CF1*Rkms)]
NumNICER_KC = [i.sum() for i in (Z_KC1*Rkms)]
```

```python
fig, ax1 = plt.subplots(nrows=1,ncols=1,figsize=(3,3),dpi=200)

ax1.plot(NMA_GF[0],NumNICER_GF,linestyle='-',label=r'$\Delta_{GF}$')
ax1.plot(NMA_QL[0],NumNICER_QL,linestyle='-',label=r'$\Delta_{QL}$')
ax1.plot(NMA_PG[0],NumNICER_PG,linestyle='-',label=r'$\Delta_{PG}$')
ax1.plot(NMA_CF[0],NumNICER_CF,linestyle='-',label=r'$\Delta_{CF}$')
ax1.plot(NMA_KC[0],NumNICER_KC,linestyle='-',label=r'$\Delta_{KC}$')

ax1.set_xbound(0,NMA_KC[0][-1])
ax1.set_ybound(0,11)
ax1.set_xlabel(r'$\textrm{Anisotropy factor}\,\left(C\right)$',fontsize=10)
ax1.set_ylabel(r'$\textrm{Number of NICER acceptable models}$',fontsize=10)

ax2 = inset_axes(ax1,width="40%",height="40%",bbox_to_anchor=(.5,.5,1.15,1.15),
                   bbox_transform=ax1.transAxes, loc=3)

ax2.plot(NMA_GF[0],NumNICER_GF,linestyle='-',label=r'$\Delta_{GF}$')
ax2.plot(NMA_QL[0],NumNICER_QL,linestyle='-',label=r'$\Delta_{QL}$')
ax2.plot(NMA_PG[0],NumNICER_PG,linestyle='-',label=r'$\Delta_{PG}$')
ax2.plot(NMA_CF[0],NumNICER_CF,linestyle='-',label=r'$\Delta_{CF}$')
ax2.plot(NMA_KC[0],NumNICER_KC,linestyle='-',label=r'$\Delta_{KC}$')
ax2.set_xbound(0,0.075)
ax2.set_ybound(0,10)
ax2.tick_params(axis='both',which='major',labelsize=6)
ax1.text(-0.053772,7.85,r'$\bullet$',fontsize=9)
ax2.text(-0.00195,7.7,r'$\bullet$',fontsize=9)
ax1.text(0.05,7.5,r'$\textrm{\textbf{Isotropy}}$',fontsize=7)
ax2.text(0.002,7.15,r'$\textrm{\textbf{Isotropy}}$',fontsize=5)

ax1.legend(ncol=1,loc=4,fontsize=6,edgecolor='k',fancybox=True,shadow=False,framealpha=0.15)

plt.show()
```
